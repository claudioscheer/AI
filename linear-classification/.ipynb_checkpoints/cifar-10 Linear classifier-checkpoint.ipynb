```python
# Load the cifar-10 dataset
import numpy as np
import cPickle

dataset= cPickle.load(open("cifar-10-batches-py\\data_batch_2",'rb'))
X = np.array(dataset['data'].reshape(10000, 3, 32, 32).transpose(0,2,3,1).astype("uint8"))
Y = np.array(dataset['labels'])
labels = cPickle.load(open("cifar-10-batches-py\\batches.meta",'rb'))['label_names']
```

```python
# Show a few random images....
from matplotlib import pyplot as plt
fig, axes = plt.subplots(nrows=2, ncols=10,figsize=(12,3),subplot_kw={'xticks': [], 'yticks': []})
for ax in axes.flat:
    index=np.random.randint(10000)
    ax.imshow(X[index], interpolation='nearest')
    ax.set_title(labels[Y[index]])
```

```python
def normalize(X):
    mean_image = np.mean(X, axis=0)
    nX=X.astype("float")-mean_image
    return nX

def svm_loss_vectorized(W,X,Y,reg,delta):
    """
    Inputs have dimension D, there are C classes, and we operate on minibatches
    of N examples.

    Inputs:
        - W: A numpy array of shape (D, C) containing weights and bias.
        - X:  numpy array of shape (N, D) containing a minibatch of data, already normalized and bias applied.
        - Y: A numpy array of shape (N,) containing training labels; y[i] = c means
            that X[i] has label c, where 0 <= c < C.
        - reg: (float) regularization strength
        - delta: (float) margin in svm

    Returns a tuple of:
        - loss as single float
        - gradient with respect to weights W; an array of same shape as W
    """
    loss = 0.0
    dW = np.zeros(W.shape) # initialize the gradient as zero
    num_train = X.shape[0]
    score=X.dot(W) #shape[10,n]
    yi_scores = score[np.arange(score.shape[0]),Y] #shape[n]
#    print "X shape:" + str(X.shape)
#    print "score shape:" + str(score.shape)
    margins = np.maximum(0, score - np.matrix(yi_scores).T + delta) #shape:[10,n]
    margins[np.arange(num_train),Y] = 0
#    print "Size of margins:" + str(margins.shape)
    loss = np.mean(np.sum(margins, axis=1))
#    print "Data loss:"+str(loss)
    loss += 0.5 * reg * np.sum(W * W)
#    print "Total loss:" +str(loss)

    binary = margins
    binary[margins > 0] = 1
    row_sum = np.sum(binary, axis=1)
#    print "Size of row_sum:" + str(row_sum.shape)
    binary[ np.arange(num_train),Y] = -row_sum.T
#    print "binary Shape:" + str(binary.shape)
    dW = np.dot(X.T, binary)
#    print "dW shape:" + str(dW.shape)

    # Average
    dW /= num_train
    # Regularize
#    dW += reg*W
    return loss, dW

def show_W(W,labels):
    W=(W-np.amin(W))/(np.amax(W)-np.amin(W))*255
    fig, axes = plt.subplots(nrows=1, ncols=10,figsize=(12,3),subplot_kw={'xticks': [], 'yticks': []})
    for idx, ax  in enumerate(axes.flat):
        ax.imshow(W.T[idx][:-1].reshape(32, 32,3 ), interpolation='nearest')
        ax.set_title(labels[idx])
```

```python
# Initialize hyperparameters...
W=np.random.uniform(low=-0.1, high=0.1,size=10*3073).reshape(3073,10)
delta=2 # margin in SVM loss
reg=1e-2 # hyperparameter for regularization loss (lambda) 
training_set =100 # number of images the loss is calculated for before gradient descent applied
learning_rate=1e-5
```

```python
#train the classifier
losses=[]
for i in range(0,100000):
    idx=np.random.randint(10000,size=training_set)
    X_train=np.hstack([normalize(X[idx]).reshape(training_set,3072),np.ones((training_set,1))]) #shape:[n,3073]
    Y_train=Y[idx] #shape:[n]
    loss,dW=svm_loss_vectorized(W,X_train,Y_train,reg,delta)
    W += -1 * learning_rate * dW
    losses.append(loss)
print "Iteration:"+str(i)+" Loss:"+str(loss)
plt.plot(losses)
plt.show()
show_W(W,labels)
    
```

```python
#Test trained classifier
test_set=100
idx=np.random.randint(10000,size=test_set)
X_train=np.hstack([normalize(X[idx]).reshape(test_set,3072),np.ones((test_set,1))]) #shape:[n,3073]
Y_train=Y[idx] #shape:[n]
score=X_train.dot(W)
classified_image=np.argmax(score,axis=1)
result = np.sum(classified_image==Y_train)
print result
```
