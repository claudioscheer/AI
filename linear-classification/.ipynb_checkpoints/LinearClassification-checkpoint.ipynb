# Linear Classification

Linear classification is about performing image classification using a function that maps the main features of a data set into a linear function. An example is shown in the image below.

The performance of this classification method is poor, but it helps to understand some of the main concepts. A good image classification algorithm uses a non-linear classification model.

## Code 

Ensures that the random numbers will always be the same:

```python
import numpy as np

np.random.seed(1)
```

Load dataset batches:

```python
import pickle

def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

Function to reshape and plot an image from the dataset:

```python
import matplotlib.pyplot as plt


def reshape_plot_cifar_image(image, label):
    reshaped_image = np.transpose(np.reshape(image, (3, 32, 32)), (1, 2, 0))
    plt.imshow(reshaped_image)
    plt.title(label)
```

Load a batch from the dataset:

```python
batches_meta = unpickle('data/batches.meta')
images_per_batch = batches_meta[b'num_cases_per_batch']
label_names = batches_meta[b'label_names']
data_1 = unpickle('data/data_batch_1')
```

Load an image to test the dataset:

```python
test_index = 5

test_image = data_1[b'data'][test_index]
test_label = label_names[data_1[b'labels'][test_index]]

reshape_plot_cifar_image(test_image, test_label.decode('UTF-8'))
```

### Train a linear model with only one image and random weights

Function to predict the output:

```python
def predict(W, x, b):
    return np.dot(W, x) + b
```


Create initial random weight and bias:

```python
np.random.seed(0)
# The weights that the model will learn to best fit the linear classification.
weigths = np.random.rand(10, 3072)
# The probability that an image belongs to a class. Helps to generalize the model.
bias = np.random.rand(10)
image = data_1[b'data'][0]
scores = predict(weigths, image, bias)
```

High Loss:

```python
def high_loss():
    for d in data_1[b'data']:
        return d
```
